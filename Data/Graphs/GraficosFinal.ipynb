```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
```

```python
TOI_pin=np.loadtxt("TOI.csv",delimiter=",",skiprows=0,usecols=[0])
TOI_pflow=np.loadtxt("TOI.csv",delimiter=",",skiprows=0,usecols=[1])
TOI_pfhigh=np.loadtxt("TOI.csv",delimiter=",",skiprows=0,usecols=[2])
TOI_p3low=np.loadtxt("TOI.csv",delimiter=",",skiprows=0,usecols=[3])
TOI_p3high=np.loadtxt("TOI.csv",delimiter=",",skiprows=0,usecols=[4])
```

```python
plt.figure()
plt.grid()
plt.xticks(np.arange(-13, 20, step=2))
plt.yticks(np.arange(-50, 50, step=10))
plt.xlabel('Potencia de entrada [dBm]')
plt.ylabel('Potencia de salida [dBm]')
plt.plot(TOI_pin,TOI_pflow,linewidth=3,label='Frecuencia fundamental')
plt.plot(TOI_pin,TOI_p3low,linewidth=3,label='Producto de 3er orden')
#plt.plot(TOI_pin,TOI_p3high)
TOI_pinext=np.linspace(-13,15,100)
plt.plot(TOI_pinext,0.96710*TOI_pinext+14.402,linestyle='dashed',label='Ajuste pendiente 1')
plt.plot(TOI_pinext,3.3653*TOI_pinext-10.1389,linestyle='dashed',label='Ajuste pendiente 3')
#plt.plot(TOI_pinext,3.369357143*TOI_pinext-8.385535714)
ip3in=10.233
ip3out=24.2985
plt.scatter(ip3in,ip3out,color='black',label='$IP_3$')
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('Amp2Tone.pdf',bbox_inches='tight')
```

```python
FiltImpedances_freq=np.loadtxt("FiltImpedances.csv",delimiter=",",skiprows=0,usecols=[0])
FiltImpedances_reZin=np.loadtxt("FiltImpedances.csv",delimiter=",",skiprows=0,usecols=[1])
FiltImpedances_imZin=np.loadtxt("FiltImpedances.csv",delimiter=",",skiprows=0,usecols=[2])
FiltImpedances_reZout=np.loadtxt("FiltImpedances.csv",delimiter=",",skiprows=0,usecols=[3])
FiltImpedances_imZout=np.loadtxt("FiltImpedances.csv",delimiter=",",skiprows=0,usecols=[4])
```
